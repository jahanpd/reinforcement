```python
import numpy as np
import seaborn as sns

values = np.loadtxt("prog.out",delimiter=",")
steps = 1237
total = int(len(values)/steps)
values = values.reshape(total,steps)
# plot avg and best case alone
# plot examples of 5 values across epochs
for i in range(5):
    sns.lineplot(np.arange(0,len(values[i]),1),values[i])
        
```

```python
idx = np.round(np.linspace(2,len(values)-1,3)).astype(int)
idx = np.append(np.array([0]),idx)
print(idx)
colours = ["red", "blue", "orange", "green", "purple", "black"]
for i in range(len(idx)):
        sns.lineplot(np.arange(0,len(values[idx[i]]),1),values[idx[i]], color=colours[i])
```
